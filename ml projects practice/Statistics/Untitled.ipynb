```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
sns.kdeplot(df["sepal_length"])
```

```python
temp=sorted(df['sepal_length'].tolist())
```

```python
y_quant=[]

for i in range(1,100):
    y_quant.append(np.percentile(temp,i))
```

```python
samples= np.random.normal(loc=0,scale=1,size=1000)#mean,std, datapoint
```

```python
x_quant=[]

for i in range(1,100):
    x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
# using statsmodel

import statsmodels.api as sm
import matplotlib.pyplot as plt

fig=sm.qqplot(df["sepal_length"],line='45',fit=True)

plt.title("QQ plot")
plt.xlabel("Therotical quantiles")
plt.ylabel("sample quantile")

plt.show()
```

# uniform distribution

```python
uniform= np.random.uniform(low=0,high=1,size=1000)
```

```python
sns.histplot(uniform)
```

```python
import scipy.stats as stats

param=stats.uniform.fit(uniform)
dist=stats.uniform(loc=param[0],scale=param[1])

fig=sm.qqplot(uniform,dist=dist,line='45')
plt.title("QQ plot with uniform distrubtion")
plt.xlabel("Therotical quantiles")
plt.ylabel("sample quantile")

plt.show()
```
