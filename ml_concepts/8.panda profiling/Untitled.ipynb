```python
import os
import json
import pandas as pd

# Define the path to the directory containing your JSON files
json_dir = '/Users/nikhil/Desktop/Research_project/Research_project/ipl_json'

match_data = []
delivery_data = []

# Iterate over each JSON file in the directory
for file_name in os.listdir(json_dir):
    if file_name.endswith('.json'):
        file_path = os.path.join(json_dir, file_name)
        with open(file_path, 'r') as file:
            data = json.load(file)

            # Extract general match details
            match_info = data.get('info', {})
            meta_info = data.get('meta', {})
            match_id = str(meta_info.get('data_version'))  # Using 'data_version' as a placeholder for match_id

            match_row = {
                'match_id': match_id,
                'season': 'IPL-2017',  # Placeholder, adjust as necessary
                'city': match_info.get('city', None),
                'date': match_info.get('dates', [None])[0],
                'team1': match_info.get('teams', [None, None])[0],
                'team2': match_info.get('teams', [None, None])[1],
                'toss_winner': match_info.get('toss', {}).get('winner', None),
                'toss_decision': match_info.get('toss', {}).get('decision', None),
                'result': match_info.get('outcome', {}).get('result', 'normal'),
                'dl_applied': match_info.get('outcome', {}).get('method', None) == 'D/L',
                'winner': match_info.get('outcome', {}).get('winner', None),
                'win_by_runs': match_info.get('outcome', {}).get('by', {}).get('runs', 0),
                'win_by_wickets': match_info.get('outcome', {}).get('by', {}).get('wickets', 0),
                'player_of_match': match_info.get('player_of_match', [None])[0],
                'venue': match_info.get('venue', None),
                'umpire1': match_info.get('officials', {}).get('umpires', [None, None])[0],
                'umpire2': match_info.get('officials', {}).get('umpires', [None, None])[1],
                'umpire3': match_info.get('officials', {}).get('tv_umpires', [None, None])[0]
            }
            match_data.append(match_row)

            # Extract delivery data
            inning_number = 0
            for inning in data.get('innings', []):
                inning_number += 1  # Manually incrementing inning number based on the order they appear
                team = inning.get('team')
                for over in inning.get('overs', []):
                    for delivery in over.get('deliveries', []):
                        delivery_data.append({
                            'match_id': match_id,
                            'inning': inning_number,
                            'batting_team': team,
                            'over': over.get('over'),
                            'ball': delivery.get('ball', 1),  # Assuming ball number is provided or default to 1
                            'batsman': delivery.get('batter'),
                            'non_striker': delivery.get('non_striker'),
                            'bowler': delivery.get('bowler'),
                            'runs_total': delivery.get('runs', {}).get('total', 0),
                            'extras_total': delivery.get('runs', {}).get('extras', 0)  # Include extras if needed
                        })

# Convert lists to pandas DataFrames for easier manipulation and viewing
match_df = pd.DataFrame(match_data)
delivery_df = pd.DataFrame(delivery_data)
```
