```python
total_score_df=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
match_df=match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df.head()
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df=match_df[['match_id','city','date','toss_winner','toss_decision','winner','team1','team2','total_runs','venue']]
```

```python
match_df.head(5)
```

```python
# First, create a helper function to label batting teams per inning
def label_batting_teams(row, inning):
    if row['inning1_team'] == row['team1']:
        return row['team1'] if inning == 1 else row['team2']
    else:
        return row['team2'] if inning == 1 else row['team1']

# Modify pivot_score_df to include team names
pivot_score_df = total_score_df.pivot_table(index='match_id', columns='inning', values='total_runs', aggfunc='first').reset_index()
pivot_score_df.columns = ['match_id', 'inning1_runs', 'inning2_runs', 'inning3_runs', 'inning4_runs', 'inning5_runs']

# Merge pivot_score_df with delivery to get team names for each inning
inning_teams = delivery[['match_id', 'inning', 'batting_team']].drop_duplicates().pivot(index='match_id', columns='inning', values='batting_team').reset_index()
inning_teams.columns = ['match_id', 'inning1_team', 'inning2_team', 'inning3_team', 'inning4_team', 'inning5_team']

# Merge the inning teams into pivot_score_df
pivot_score_df = pivot_score_df.merge(inning_teams, on='match_id', how='left')

# Merge this modified pivot_score_df with match_df
match_df = match_df.merge(pivot_score_df[['match_id', 'inning1_runs', 'inning2_runs', 'inning1_team', 'inning2_team']], on='match_id', how='left')

# Optionally, clean up the DataFrame by adjusting column orders or dropping unnecessary ones
match_df = match_df[['match_id', 'city', 'date', 'team1', 'team2', 'toss_winner', 'toss_decision', 'winner', 'venue', 'inning1_team', 'inning1_runs', 'inning2_team', 'inning2_runs']]


```
