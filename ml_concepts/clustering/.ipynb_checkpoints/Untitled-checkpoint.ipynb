```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
```

```python
n_samples=300
n_features=2
centres=4

#data genration

X,y= make_blobs(n_samples=n_samples,n_features=n_features,centers=centres,cluster_std=1.5,random_state=42)

plt.scatter(X[:,0],X[:,1],s=50,cmap='virdis')
plt.title("genrate data set for k mean clustering")
plt.xlabel("feature 1")
plt.ylabel(" feature 2")
plt.show()
```

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_init=10,n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
kmeans=KMeans(n_init=10,n_clusters=4)
```

```python
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=60,c='red',label='Cluster1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=60,c='blue',label='Cluster2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=60,c='green',label='Cluster3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=60,c='yellow',label='Cluster4')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='centroids')

plt.legend()
plt.show()
```
