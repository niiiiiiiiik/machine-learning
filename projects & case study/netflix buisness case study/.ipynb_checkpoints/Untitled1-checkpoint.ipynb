# Netflix Content Strategy Analysis


## Introduction
Netflix, as a leading streaming platform, possesses a vast and diverse content library. This analysis aims to leverage data-driven insights to guide content acquisition, production strategies, and optimize release schedules across global markets.


## Problem Statement
The objective is to analyze the existing content database to inform strategic decisions on:
1. The type of content to prioritize for production or acquisition.
2. Optimal release strategies in different global markets.
3. Preferences for content genres and successful directors across various demographics.


## Analysis Objectives
- **Content Type Analysis:** Determine the prevalence and trends of movies vs TV shows and analyse content popularity in specific regions.
- **Geographic Content Strategy:** Identify major content-producing countries and potential markets for expansion.
- **Temporal Release Strategy:** Recommend optimal times for releasing new content to maximize engagement.
- **Genre and Director Focus:** Evaluate popular genres and directors to guide future productions and collaborations.
- **Viewer Preferences:** Assess viewer preferences through ratings and viewership data to identify potentially underserved niches.

## Handling Missing Data in the Netflix Dataset

### Overview
The dataset comprises several fields with missing values, which could impact our content strategy analysis. Here's a summary of the missing data:

- **Director:** 2,634 missing entries - affects analysis of directorial impact.
- **Cast:** 825 missing entries - limits evaluation of cast influence on success.
- **Country:** 831 missing entries - restricts understanding of geographical content preferences.
- **Date Added:** 10 missing entries - impacts assessment of content lifecycle on the platform.
- **Rating:** 4 missing entries - minor effect but essential for understanding viewer demographics.
- **Duration:** 3 missing entries - slightly affects content engagement analysis.


Addressing these gaps will provide more robust insights into Netflixâ€™s content strategy, thus informing decisions on content development, acquisition, and scheduling.


```python
df['rating'] = df['rating'].fillna('0')
df['director']=df['director'].fillna('Unknown Director')
df['cast']=df['cast'].fillna("Unknown Actors")
df['country']=df['country'].fillna('Unknown')
df['duration']=df['duration'].fillna('Unknown')
```

```python
# Convert to datetime if not already
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Fill with the mode (most common date)
mode_date = df['date_added'].mode()[0]
df['date_added'].fillna(mode_date, inplace=True)
```

```python
df.isna().sum()
```

```python
## add new features in the dataset
df["date_added"] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

df['season_count'] = df.apply(lambda x : x['duration'].split(" ")[0] if "Season" in x['duration'] else "", axis = 1)
df['duration'] = df.apply(lambda x : x['duration'].split(" ")[0] if "Season" not in x['duration'] else "", axis = 1)
df.head()
```



### Data Imputation Justifications
- **Rating**: Missing ratings filled with '0' to denote unrated content, maintaining data integrity for analysis.
- **Director**: 'Unknown Director' placeholder used for missing entries allows for inclusive data analysis without data loss.
- **Cast**: 'Unknown Actors' assigned to missing cast details ensure all titles are usable in cast-related analyses.
- **Country**: Inputting 'Unknown' for missing country data allows for geographical analyses without excluding any records.
- **Duration**: Replacing missing durations with 'Unknown' keeps all entries in duration-based analyses.
- **Date Added**: The mode date used for missing 'date_added' entries preserves the timeline for release trend analysis.


- **Year and Month Added**: Extracted from 'date_added' to analyze content release patterns and seasonal effects.
- **Duration and Season Count**:
  - **Movies**: Numeric duration values extracted to facilitate content length analysis.
  - **TV Shows**: Season count extracted for analyzing series longevity and renewal viability.


## 1. Content Type on Netflix 

```python
type_counts = df['type'].value_counts().reset_index()
type_counts = type_counts.rename(columns={'type': 'count', 'index': 'type'})


fig, ax = plt.subplots()
ax.pie(type_counts['count'], labels=type_counts['type'], autopct='%1.1f%%', startangle=90, colors=["#6ad49b", "#a678de"])
ax.axis('equal')  
plt.title("Distribution of Netflix Titles by Type")
plt.show()
```

Distribution of Netflix Titles by Type: This pie chart illustrates the proportion of content types available on Netflix, categorised into Movies and TV Shows. From the chart, it is evident that Movies constitute the majority, accounting for approximately 69.6% of the content, whereas TV Shows make up about 30.4%. This indicates a more substantial movie selection, which could be a strategic content distribution by Netflix to cater to a broader audience preference for films.

## 2. Growth in content over the years 

```python

d1 = df[df["type"] == "TV Show"]
d2 = df[df["type"] == "Movie"]


vc1 = d1['year_added'].value_counts().reset_index()
vc1 = vc1.rename(columns={'year_added': 'count', 'index': 'year_added'}).sort_values('year_added')

vc2 = d2['year_added'].value_counts().reset_index()
vc2 = vc2.rename(columns={'year_added': 'count', 'index': 'year_added'}).sort_values('year_added')


data = {
    "Year": vc1['year_added'].tolist() + vc2['year_added'].tolist(),
    "Count": vc1['count'].tolist() + vc2['count'].tolist(),
    "Type": ["TV Show"] * len(vc1) + ["Movie"] * len(vc2)
}
df_plot = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_plot, x="Year", y="Count", hue="Type", style="Type", markers=True, dashes=False)
plt.title("Growth of Netflix Content Over the Years")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles Added")
plt.legend(title="Content Type")
plt.show()
```

Growth of Netflix Content Over the Years: The line graph presents the annual growth in the number of titles added to Netflix, differentiated by content type (Movies and TV Shows). From 2008 to 2020, there is a noticeable trend where the number of both Movies and TV Shows added increases substantially. The sharp rise starting around 2016 suggests an aggressive expansion of Netflix's library, coinciding with its push into new geographic markets and increased investment in original content. Notably, the growth peaks in 2018 for TV Shows, followed by a slight decline, possibly reflecting market saturation or strategic shifts in content acquisition and production.

The bar chart presents the original release year of Netflix content over the last decade, categorized by content type: TV Shows and Movies. Each bar represents the number of titles released in a given year, with TV Shows in green and Movies in purple.

From the chart:

TV Shows: There is a clear upward trend in the number of TV shows added to Netflix, particularly from 2016 onwards. The peak occurs in 2018, with a slight decrease thereafter but still maintaining high numbers through 2021. This surge reflects Netflixâ€™s increasing focus on expanding its original series offerings.

Movies: The addition of movies has also seen a general increase over the decade, though not as pronounced as TV shows. The number of movies added each year shows variability, with some years like 2015 and 2018 seeing significant increases.

## Analysis of Movie Durations on Netflix Over the Last 15 Years




### Key Observations

- **Variability**: The interquartile range (IQR), represented by the box's length, shows the middle 50% of movie durations range from approximately 90 to 120 minutes.
- **Outliers**: The plot highlights several outliers, indicating movies that are significantly shorter or longer than the typical durations. These are represented by the individual points beyond the whiskers.

## Distribution of Ratings for TV Shows and Movies on Netflix

```python
col = "rating"

vc1 = d1[col].value_counts().reset_index()
vc1 = vc1.rename(columns = {col : "count", "index" : col})
vc1['percent'] = vc1['count'].apply(lambda x : 100*x/sum(vc1['count']))
vc1 = vc1.sort_values(col)

vc2 = d2[col].value_counts().reset_index()
vc2 = vc2.rename(columns = {col : "count", "index" : col})
vc2['percent'] = vc2['count'].apply(lambda x : 100*x/sum(vc2['count']))
vc2 = vc2.sort_values(col)


vc1['Type'] = 'TV Shows'
vc2['Type'] = 'Movies'
combined_data = pd.concat([vc1, vc2])
combined_data['rating'] = combined_data['rating'].replace(r'\d+ min', "NA", regex=True)
combined_data['rating']=combined_data['rating'].replace('0','NA')


plt.figure(figsize=(12, 6))
sns.barplot(data=combined_data, x='rating', y='count', hue='Type', palette=["#a678de", "#6ad49b"])
plt.title('Distribution of Ratings for TV Shows and Movies')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

## Analysis of Most Frequent Cast Members in Movies Across Selected Countries on Netflix

```python


def prepare_country_data(df, country, flag="movie"):
    df["from_us"] = df['country'].fillna("").apply(lambda x : 1 if country.lower() in x.lower() else 0)
    small = df[df["from_us"] == 1]
    
    if flag == "movie":
        small = small[small["duration"] != ""]
    else:
        small = small[small["season_count"] != ""]
    
    cast = ", ".join(small['cast'].fillna("")).split(", ")
    tags = Counter(cast).most_common(25)
    tags = [_ for _ in tags if _[0] != ""]

    labels = [_[0] for _ in tags][::-1]
    values = [_[1] for _ in tags][::-1]

    return labels, values



fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 30))  # Increased figure size
countries = ["United States", "India", "United Kingdom", "Canada", "Spain", "Japan"]
palette = sns.color_palette("hsv", len(countries))  # Unique color for each country

for ax, country, color in zip(axes.flatten(), countries, palette):
    labels, values = prepare_country_data(df, country, flag="movie")
    if labels and values:  # Check if there is data to plot
        sns.barplot(x=values, y=labels, ax=ax, color=color)
        ax.set_title(f'Most Frequent Cast in {country}', fontsize=16)  # Increased font size
        ax.set_xlabel('Counts', fontsize=14)
        ax.set_ylabel('')
        ax.tick_params(labelsize=12)  # Larger tick labels
        ax.grid(True)  # Adding grid lines for better readability
    else:
        ax.set_title(f'No data available for {country}', fontsize=16)

plt.tight_layout()
plt.show()
```

### Key Observations
- **United States**: Features a diverse range of well-known actors such as Samuel L. Jackson and Adam Sandler, indicating their broad appeal in American films.
- **India**: Dominated by Bollywood heavyweights like Shah Rukh Khan and Kareena Kapoor, reflecting the stars' significant roles in the Indian cinema sector.
- **United Kingdom**: Showcases a mix of classic and contemporary actors such as Jude Law and Johnny Depp, highlighting the cross-generational appeal of British cinema.

This cast popularity can guide Netflix in future productions and marketing strategies, particularly in tailoring content to match viewer preferences in different regions. Understanding these dynamics is crucial for content localization and global content strategy adaptation.

The data also underscores the role of star power in driving the popularity and viewership of films, influencing Netflix's decisions on which projects to greenlight, which actors to cast, and how to strategically market content in various locales.

## Analysis of Top Directors in Movies and TV Shows

```python
def prepare_director_data(df, country, content_type):
    df["from_country"] = df['country'].fillna("").apply(lambda x: 1 if country.lower() in x.lower() else 0)
    small = df[df["from_country"] == 1]
    
    if content_type == "movie":
        small = small[(small["type"] == "Movie") & (small["director"].notna())]
    else:
        small = small[(small["type"] == "TV Show") & (small["director"].notna())]

    directors = ", ".join(small['director']).split(", ")
    counter = Counter(directors).most_common(25)
    counter = [_ for _ in counter if _[0] != ""]

    labels = [_[0] for _ in counter][::-1]
    values = [_[1] for _ in counter][::-1]
    return labels, values


countries = ["United States", "United Kingdom", "India"]
types = ["movie", "tv_shows"]


num_plots = len(countries) * len(types)
num_rows = (num_plots + 1) // 2  
fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(20, 10 * num_rows))  
axes = axes.flatten()  

plot_index = 0
for country in countries:
    for content_type in types:
        labels, values = prepare_director_data(df, country, content_type)
        if labels and values:  # Check if data is available
            sns.barplot(x=values, y=labels, ax=axes[plot_index], color="#a678de")
            axes[plot_index].set_title(f'Top Directors in {country} for {content_type}')
            axes[plot_index].set_xlabel('Counts')
            axes[plot_index].set_ylabel('')
        else:
            axes[plot_index].set_title(f'No data available for {country} {content_type}')
        axes[plot_index].tick_params(axis='y', labelsize=12) 
        axes[plot_index].grid(True)
        plot_index += 1


if num_plots % 2 != 0:
    axes[-1].set_visible(False)

plt.tight_layout()
plt.show()
```

### Insights

1. **United States**:
   - **Movies**: Dominated by directors known for their distinctive filmmaking styles such as Steven Spielberg, Quentin Tarantino, and Martin Scorsese.
   - **TV Shows**: Includes a variety of directors like Tony Callinguwood and Tyler Russell, reflecting the diversity in American television production.

2. **United Kingdom**:
   - **Movies**: Features directors like Ben Wheatley and Christopher Nolan, known for their unique visual and narrative styles.
   - **TV Shows**: Showcases directors such as Rachel Bell and Jakob Verbruggen, who are pivotal in British TV series production.

3. **India**:
   - **Movies**: Includes directors like Vishal Bhardwaj and Zoya Akhtar, who are renowned for their storytelling in Indian cinema.
   - **TV Shows**: Captures directors such as Vetri Maaran and Vignesh Shivan, who have made significant contributions to Indian television.


## Genre Popularity Trends on Netflix

```python
df['genres'] = df['listed_in'].apply(lambda x: x.split(', '))
df = df.explode('genres')
genre_trends = df.groupby(['year_added', 'genres']).size().unstack(fill_value=0).cumsum()


plt.figure(figsize=(14, 8))  
genre_trends.plot(kind='line', marker='o', figsize=(14, 8), colormap='tab20')  
plt.title('Genre Popularity Trends Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Titles', fontsize=14)
plt.grid(True)
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.tight_layout()  
plt.show()
```

### Observations:

- **Rapid Growth in Diverse Genres**: Starting around 2016, there is a noticeable surge in a variety of genres. Notably, genres such as Documentaries, Dramas, and International Movies show significant growth, indicating an expansion in Netflix's content diversity.
- **Steady Increase in Niche Genres**: Genres like Anime, Sci-Fi & Fantasy, and Horror have seen steady increases, reflecting a strategic broadening to cater to niche audiences.
- **Rise of International Content**: The sharp upward trajectories for International Movies and TV Shows highlight Netflix's increased focus on global markets, tailoring content to cater to international viewers.



### Regional Content Preferences in Netflix Distribution

```python
df['genres'] = df['listed_in'].apply(lambda x: x.split(', '))
df_exploded = df.explode('genres')


def get_genre_counts(country):
    filtered_genres = df_exploded[df_exploded['country'].str.contains(country, na=False)]
    return filtered_genres['genres'].value_counts().reset_index().head(10).rename(columns={'index': 'Genre', 'genres': 'Count'})


us_genre_counts = get_genre_counts("United States")
uk_genre_counts = get_genre_counts("United Kingdom")
india_genre_counts = get_genre_counts("India")
japan_genre_counts = get_genre_counts("Japan")


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 16))  # Adjust size as necessary
axes = axes.flatten()


sns.barplot(data=us_genre_counts, x='Count', y='Genre', ax=axes[0], palette='Blues_d')
axes[0].set_title('Top 10 Genres in the United States')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Genre')

sns.barplot(data=uk_genre_counts, x='Count', y='Genre', ax=axes[1], palette='Greens_d')
axes[1].set_title('Top 10 Genres in the United Kingdom')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Genre')

sns.barplot(data=india_genre_counts, x='Count', y='Genre', ax=axes[2], palette='Reds_d')
axes[2].set_title('Top 10 Genres in India')
axes[2].set_xlabel('Count')
axes[2].set_ylabel('Genre')

sns.barplot(data=japan_genre_counts, x='Count', y='Genre', ax=axes[3], palette='Purples_d')
axes[3].set_title('Top 10 Genres in Japan')
axes[3].set_xlabel('Count')
axes[3].set_ylabel('Genre')

plt.tight_layout()
plt.show()
```

### Business Insights:
Based on the comprehensive analysis of the Netflix dataset spanning genres, content types, and geographical distributions, the following key patterns and insights have been observed:

2. **Regional Content Preferences**: The data shows significant regional disparities in content production and popularity. For instance, the US predominantly produces TV shows and movies in genres like comedy and action, whereas India focuses more on dramas and romantic movies, reflecting cultural preferences that could influence regional marketing strategies.

3. **Impact of Directors**: Certain directors appear frequently within specific regions, indicating their significant influence on local content popularity. 4. **Viewer Ratings and Content Type**: There's a clear preference for TV-MA and TV-14 rated content, reflecting a viewer base that skews towards mature audiences. This suggests that edgier, more adult-themed content might be more successful in terms of engagement and viewership.

5. **Seasonal Viewing Patterns**: Analysis of content added over the years shows peaks typically around December and the summer months, aligning with holiday seasons when viewers have more leisure time, indicating that these periods may be optimal for releasing high-profile or highly anticipated content to maximize viewership.

6. **Viewer Engagement Trends**: The growth in the number of titles over the years, particularly post-2014, aligns with the global expansion of Netflix into new markets, showing how geographical expansion is likely correlated with increased content production and diversity to cater to a broader audience.

### Recommendations:
Based on the insights derived from the data analysis, the following actionable recommendations are proposed to enhance Netflix's content strategy:

1. **Invest in Growing Genres**: Continue to invest in and expand international films and documentaries. These genres have shown significant growth and appeal to a broad audience, suggesting potential for high engagement and subscriber growth.

2. **Tailor Content by Region**: Develop and promote content that aligns with regional tastes and cultural preferences, such as more drama and romance in India and action in the US. Localising content can lead to better viewer satisfaction and retention.

3. **Leverage Popular Directors**: Forge stronger ties and create exclusive deals with popular directors such as David Attenborough or Anurag Kashyap to tap into their fanbase and pull new subscribers.

4. **Optimize Release Schedules**: Schedule the release of new seasons and high-profile titles during peak viewing times identified from the analysis, such as during holidays and special events to capture the maximum audience.

5. **Focus on Quality and Ratings**: Enhance quality control measures to ensure content not only attracts but retains audience interest. High-rated content tends to retain viewers longer, contributing to sustained subscription revenue.

6. **Expand Viewer Base**: Given the popularity of mature content, consider developing more series and films with TV-MA and TV-14 ratings while also balancing family-oriented content like children's shows and family movies to cater to all demographics.
